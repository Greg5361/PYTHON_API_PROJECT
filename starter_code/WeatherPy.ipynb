## Generate Cities List

```python
from api_keys import weather_api_key, g_key
import matplotlib.pyplot as plt
import pandas as pd
import requests as rq
import time
from scipy.stats import linregress
from citipy import citipy
import numpy as np
output = '../output_data/cities.csv'
```

```python
lats = np.random.uniform(low = -90, high = 90, size=1500)
lngs = np.random.uniform(low = -180, high = 180, size=1500)
lats_lngs = zip(lats,lngs)
```

```python
cities = []
for lat_lng in lats_lngs:
    city = citipy.nearest_city(lat_lng[0],lat_lng[1]).city_name
    if city not in cities:
        cities.append(city)
len(cities)
```

### Perform API Calls
* Perform a weather check on each city using a series of successive API calls.
* Include a print log of each city as it'sbeing processed (with the city number and city name).


```python
weather = []
print('\nBeginning Data Retrieval\n-----------------------------')
rcd = 0
sets = 1

for city in cities:
    rcd+=1
    print(f'Processing Record {rcd} of Set {sets} | {city}')
    
    if rcd == 50:
        rcd = 0
        sets+=1
    
    try:
        data = {}
        url = f'https://api.openweathermap.org/data/2.5/weather?q={city}&appid={weather_api_key}&units=imperial'
        res = rq.get(url).json()
        data['City'] = city
        data['Lat'] = res['coord']['lat']
        data['Lng'] = res['coord']['lon']
        data['Temp'] = res['main']['temp_max']
        data['Humidity'] = res['main']['humidity']
        data['Cloudiness'] = res['clouds']['all']
        data['Wind Speed'] = res['wind']['speed']
        data['Country'] = res['sys']['country']
        data['Date'] = res['dt']
        weather.append(data)
    except:
        print('City not found. Skipping...') 

print('\n-----------------------------\nData Retrieval Complete\n-----------------------------\n')
```

### Convert Raw Data to DataFrame
* Export the city data into a .csv.
* Display the DataFrame

```python
df = pd.DataFrame(weather)
df.head()
```

## Linear Regression

```python
def LineRegression(x,y,title):
    (slope,intercept,rvalue,pvalue,stderr)=linregress(x,y)
    regress=x*slope+intercept
    msg=f'Y = {slope:.2f}X + {intercept:.2f}'
    plt.scatter(x,y,linewidths=1,edgecolor='black',label='Cities',alpha=1)
    plt.plot(x,regress,'r')
    plt.annotate(msg,(x.min(),y.min()),fontsize=15,color='red')
    plt.title(f'Latitude vs {title}')
    plt.ylabel(f'{title}')
    plt.xlabel('Latitude')
    plt.show()
```
